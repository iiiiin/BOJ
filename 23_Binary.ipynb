```python
# 1920 - 수 찾기

def solution(array, target, start, end):
    while start <= end:
        mid = (start + end) // 2
        if array[mid] == target:
            return 1
        elif array[mid] > target:
            end = mid - 1
        else:
            start = mid + 1
    return 0

n = int(input())
a = sorted(list(map(int, input().split())))
m = int(input())
array = list(map(int, input().split()))

for i in range(m):
    print(solution(a, array[i], 0, n-1))
```

```python
# 10816 - 숫자 카드 2 - 이분 탐색

import sys
input = sys.stdin.readline
a = int(input())
N = sorted(map(int,input().split()))
b = int(input())
M = map(int,input().split())

def binary(n, N, start, end):
    if start > end:
        return 0
    m = (start+end)//2
    if n == N[m]:
        return N[start:end+1].count(n)
    elif n < N[m]:
        return binary(n, N, start, m-1)
    else:
        return binary(n, N, m+1, end)

n_dic = {}
for n in N:
    start = 0
    end = a - 1
    if n not in n_dic:
        n_dic[n] = binary(n, N, start, end)

print(' '.join(str(n_dic[x]) if x in n_dic else '0' for x in M ))
```

```python
# 1654 - 랜선 자르기

import sys
input = sys.stdin.readline

def solution(target,array,start,end):
    result = 0
    while start <= end:
        mid = (start + end) // 2
        total = sum([i // mid if (i // mid) > 0 else 0 for i in array])
        if total >= target:
            start = mid+1
            result = mid
        else:
            end = mid-1
    return result

k, n = map(int, input().split())
array = [int(input()) for i in range(k)]
print(solution(n,array,1,max(array)))
```

```python
# 2805 - 나무 자르기

def solution(target,array,start,end):
    result = 0
    while start <= end:
        mid = (start + end) // 2
        total = sum([i - mid if i > mid else 0 for i in array])
        if total >= target:
            start = mid+1
            result = mid
        else:
            end = mid-1
    return result

k, n = map(int, input().split())
array = list(map(int, input().split()))
print(solution(n,array,1,max(array)))
```

```python
# 2110 - 공유기 설치

def solution(target,array,start,end):
    result = 0
    while start <= end:
        mid = (start + end) // 2
        point = array[0]
        count = 1
        for i in range(1,len(array)):
            if array[i] - point >= mid:
                count += 1
                point = array[i]
        if count >= target:
            start = mid + 1
            result = mid
        else:
            end = mid - 1
    return result

n, c = map(int, input().split())
array = sorted([int(input()) for i in range(n)])
print(solution(c,array,0,array[n-1]-array[0]))
```

```python
# 1300 - K번째 수

n = int(input())
k = int(input())
start, end = 1, k
while start <= end:
    mid = (start + end) // 2
    result = 0
    for i in range(1, n+1):
        result += min(mid//i, n)
    if result >= k:
        answer = mid
        end = mid - 1
    else:
        start = mid + 1
print(answer)
```

```python
# 12015 - 가장 긴 증가하는 부분 수열 2

def solution(array,target,start,end):
    lb = 0
    while start <= end:
        mid = (start + end) // 2
        if target > array[mid]:
            start = mid + 1
        else:
            lb = mid
            end = mid - 1
    return lb

k = int(input())
array = list(map(int, input().split()))
arr = []
arr.append(array.pop(0))

for i in array:
    if arr[-1] < i:
        arr.append(i)
    else:
        lb = solution(arr,i,0,len(arr)-1)
        arr[lb] = i
print(len(arr))
```
